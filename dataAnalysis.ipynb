```python
def loadFiles(ids):
    files = []
    for ID in ids:
        doc = pd.read_csv('./'+str(ID)+'_stats.csv')
        doc = doc.select(lambda x: not re.search('Unnamed*', x), axis=1)
        doc.dropna(how='all')
        doc['id']=ID
        files.append(doc)
        doc = pd.read_csv('/Users/guillaume/Dropbox/'+str(ID)+'_stats.csv')
        doc = doc.select(lambda x: not re.search('Unnamed*', x), axis=1)
        doc.dropna(how='all')
        doc['id']=ID
        files.append(doc)
    return files
```

```python
pca = PCA(n_components=2)
pca.fit(x)
x_reduced = pca.transform(x)
print "Reduced dataset shape:", x_reduced.shape

import pylab as pl
pl.scatter(x_reduced[:, 0], x_reduced[:, 1], c=y)
# pl.scatter(x_reduced[:, 1], x_reduced[:, 2], c=y)
# pl.scatter(x_reduced[:, 0], x_reduced[:, 2], c=y)

print "Meaning of the components:"
for component in pca.components_:
    print " + ".join("%.3f x %s" % (value, name)
                     for value, name in zip(component, x.columns.values))
```

```python
from sklearn.cluster import KMeans
k_means = KMeans(n_clusters=4, random_state=0) # Fixing the RNG in kmeans
k_means.fit(x_reduced)
y_pred = k_means.predict(x_reduced)

pl.scatter(x_reduced[:, 0], x_reduced[:, 1], c=y_pred)
```
